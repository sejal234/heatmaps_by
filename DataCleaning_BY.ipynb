## Training Data: Zipcodes
Return a dataframe with Houston area zipcodes and the number of training programs that MorePathways.Org offers in those zipcodes

First, export trainings from salesforce as a CSV, ISO-8859. make sure the TL Zipcode column is in the export. Follow instructions bellow.  

```python
#replace this variable with your path to training programs
trainings = pd.read_csv('/Users/sejalgupta/Downloads/report1656557617838.csv')

#update this with where in the computer you want to save your cleaned file
save = '/Users/sejalgupta/Documents/heatmap_docs/trainings_zips.csv'

#run the cell
```

```python
#run this cell and it should save into your computer!!!
def clean_training_zips(trainings, save):
    print("Total Entries", len(trainings))
    trainings = trainings[["TL Zipcode"]]
    trainings = trainings.dropna()
    trainings['TL Zipcode'] = trainings['TL Zipcode'].astype(str).apply(lambda x: x.replace('.0',''))
    trainings["TL Zipcode"] = trainings["TL Zipcode"].astype(int)
    trainings['programs_count'] = 1
    trainings = trainings.groupby('TL Zipcode').count()[['programs_count']].reset_index()
    trainings.columns = [["ZIPCODE", "Num_Trainings"]]
    
    #comment if you do not want to save the file
    trainings.to_csv(save)
    #return trainings <- if you want to manipulate the dataframe
    
    #if you were wondering how many training programs our data has, print the cell
    print("Number of Training Programs:", int(trainings["Num_Trainings"].sum()))
    
clean_training_zips(trainings, save)
```

## Opportunity Youth: Zipcodes
Return a dataframe with Houston area zipcodes, the number of opportunity youth in the zipcode (as defined by ACS data, those 16-19 not in school and not employed). <br>
Export OY from https://data.census.gov/cedsci/ -> Table B14005 (I did Year 2020) -> Try to pull as many Houston zipcodes as possible using geos. The more, the better!

```python
#put the path to your file below
path = '/Users/sejalgupta/Downloads/ACSDT5Y2020.B14005-2022-06-30T033207.csv'

oy = pd.read_csv(path, thousands=',')
    #this dataset has all the opportunity youth in texas in 2020!!

#put the path you want to save to below
save = '/Users/sejalgupta/Documents/heatmap_docs/oy_zip_texas.csv'
```

```python
#run this cell and it should save into your computer!!!

def clean_oy_zips(oy, save):
    #if the rows are diff in future years then none of this will work
    oy = oy.drop([1,2,3,4,5,6,7,8,10,11,12,14,15,16,17,18,19,20,21,22,24,25,26,28], axis = 0)
    cols = []
    for column in oy.columns:
        oy.rename(columns = {column: column[6:11]}, inplace = True)
    oy = oy.transpose()
    oy["Opportunity Youth"] = oy[9] + oy[13] + oy[23] + oy[27]
    oy = oy[[0, "Opportunity Youth"]]
    oy = oy.reset_index()
    oy.columns = [["ZIPCODE", "Total 16-19", "Opportunity Youth"]]
    oy = oy.iloc[1: , :]
    oy["ZIPCODE"] = oy["ZIPCODE"].astype(int)
    
    #comment if you do not want to save the file
    oy.to_csv(save)
    #return oy <- if you want to manipulate the dataframe
    
clean_oy_zips(oy,save)
```
